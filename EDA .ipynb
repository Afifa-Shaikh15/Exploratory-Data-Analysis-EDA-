```python
# Titanic EDA Program (all-in-one)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Make sure titanic.csv is in the same directory)
df = pd.read_csv('C:/Users/ARSHIYA/Downloads/archive (5)/Titanic-Dataset.csv')

# 1. Basic Info
print("----- Dataset Info -----")
print(df.info())
print("\n----- First 5 Rows -----")
print(df.head())

# 2. Summary Statistics
print("\n----- Numerical Summary -----")
print(df.describe())

print("\n----- Categorical Summary -----")
print(df.describe(include='object'))

# 3. Missing Values
print("\n----- Missing Values -----")
print(df.isnull().sum())

# 4. Histograms
print("\nShowing Histogram of Age...")
plt.figure(figsize=(6, 4))
df['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 5. Boxplot - Age vs Pclass
print("\nShowing Boxplot of Age by Pclass...")
plt.figure(figsize=(6, 4))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Boxplot of Age across Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.grid(True)
plt.show()

# 6. Correlation Matrix
print("\nShowing Correlation Heatmap...")
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# 7. Pairplot (Subset of features)
print("\nShowing Pairplot...")
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']].dropna(), hue='Survived')
plt.show()

# 8. Group-wise Survival Rates
print("\n----- Survival Rate by Sex -----")
print(df.groupby('Sex')['Survived'].mean())

print("\n----- Survival Rate by Pclass -----")
print(df.groupby('Pclass')['Survived'].mean())

# 9. Skewness
print("\n----- Skewness -----")
print("Age skewness:", df['Age'].skew())
print("Fare skewness:", df['Fare'].skew())
```
